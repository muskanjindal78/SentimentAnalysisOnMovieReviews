```python
from sklearn.model_selection import train_test_split

#Randomly Splitting for unbiased evaluation
train_size = 0.8
X = train_data.drop(columns= ['sentiment']).copy()
y = train_data['sentiment']

X_train, X_rem, y_train, y_rem = train_test_split(X, y, train_size=train_size)

#now i want to divide remaining data into 50-50 for valid and test size
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=test_size)

print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
```

```python
#Data preprocessing
#Converting object data into numeric values

from sklearn.feature_extraction.text import TfidfVectorizer
# from sklearn.feature_extraction.text import CountVectorizer
# from sklearn.pipeline import Pipeline

#To impute missing values in categorical as well as numerical
from sklearn.impute import SimpleImputer

X_train = train_data.drop(columns=['sentiment']).copy()
y_train = train_data['sentiment']

X_test = test_data

#Filling null values with NaN value to make that easily fillable
X_train = X_train.fillna(np.nan)
X_test = X_test.fillna(np.nan)

#Getting all the list of columns in train and test data
X_train_cols = list(X_train.keys())
X_test_cols = list(X_test.keys())

#Getting all the types of train columns
X_train_types = dict(X_train.dtypes)

# print(X_train_cols)
# print(X_train_types)

#Filtering categorical and numerical columns
cat_dict = []
num_dict = []
bool_dict = []
for i,j in X_train_types.items():
    if j == "object":
        cat_dict.append(i)
    elif j == "float":
        num_dict.append(i)
    elif j == "bool":
        bool_dict.append(i)

#Imputing missing values
si1= SimpleImputer(missing_values=np.nan, strategy='most_frequent')
si2= SimpleImputer(missing_values=np.nan, strategy='most_frequent')

X_train1 = si1.fit_transform(X_train)
X_test1 = si2.fit_transform(X_test)

# print(X_train1)
# print(X_test1)

X_train = pd.DataFrame(X_train1, columns = X_train_cols)
X_test = pd.DataFrame(X_test1, columns = X_test_cols)

# print(X_train)    
# print(cat_dict) ['movieid', 'reviewerName', 'reviewText', 'title', 'rating', 
#'ratingContents', 'releaseDateTheaters', 'releaseDateStreaming', 'genre', 'originalLanguage',
#'director', 'boxOffice', 'distributor', 'soundType']
# print(num_dict) ['audienceScore', 'runtimeMinutes']

vectorizer = TfidfVectorizer(lowercase=True, max_features=40000)

X_train_reviewText = vectorizer.fit_transform(X_train['reviewText'])
X_test_reviewText = vectorizer.fit_transform(X_test['reviewText'])    

print(X_train_reviewText.shape)
print(X_test_reviewText.shape)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold

logit = LogisticRegression(C=5e1, solver='newton-cg', multi_class='multinomial', 
                           random_state=17, n_jobs=4)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)

cv_results = cross_val_score(logit, X_train_reviewText, y_train, 
                             cv=skf, scoring='f1_micro')
cv_results, cv_results.mean()
```

```python
logit.fit(X_train_reviewText, y_train)

test_preds = logit.predict(X_test_reviewText)
pred = pd.DataFrame(test_preds, columns=['sentiment'])
pred.index.name = 'id'
pred.to_csv('submission.csv')
```
